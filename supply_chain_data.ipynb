```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("supply_chain_data.csv")

print(df.head(10))
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
Data=df.groupby("Product type") ["Costs"].sum()
print(Data)
```

```python
Product =['Cosmetics','Haircare','Skincare']
Cost=[13366.397283,17328.862865,22229.318068]

sns.set_style("whitegrid")
plt.pie(Cost,labels=Product,autopct='%1.1f%%')
plt.axis('equal')
plt.title("Total_cost_by_Product")
plt.show()
```

```python
Data=df.groupby("Location") ["Order quantities"].sum()
print(Data)
```

```python
location =['Bangalore','Chennai','Delhi','Kolkata','Mumbai']
total_order=[769,1109,733,1228,1083]

sns.set_style("whitegrid")
plt.pie(total_order,labels=location,autopct='%1.1f%%')
plt.axis('equal')
plt.title("Total_order_location")
plt.show()
```

```python
sns.histplot(df["Inspection results"],bins=5)
plt.show()
```

```python
pd.crosstab(df["Inspection results"],df["Product type"]).plot(kind="bar")
plt.show()
```
